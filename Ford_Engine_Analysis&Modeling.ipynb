```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import normaltest, probplot
```

### Selection of Features
- Perform normality test and figure out which features do not follow normal distribution.
- Feature 'sensor8', 'sensor9','sensor10','sensor419','sensor420','sensor421' shows very high p-value so we transformed these features.

```python
train_data = pd.read_csv('./ford_train.csv', index_col = 0)
test_data = pd.read_csv('./ford_test.csv', index_col = 0)

for col in train_data.columns:
    print(col)
    _,p_val = normaltest(train_data[col])
    print(p_val)
```

```python
train_data.iloc[:,[7,8,9,418,419,420]].plot.kde(subplots=True, sharex=False, figsize = (10,20))
```

### Feature Transformations

```python
from sklearn.preprocessing import StandardScaler, RobustScaler, PowerTransformer, MinMaxScaler

std_scaler1 = StandardScaler()
std_scaler2 = StandardScaler(with_mean = False)
std_scaler3 = StandardScaler(with_mean = False, with_std = False)
r_scaler = RobustScaler()
p_scaler = PowerTransformer()
mm_scaler = MinMaxScaler()
```

```python
print(len(train_data['sensor8']))
```

```python
scaled_df = pd.DataFrame(index = range(len(train_data['sensor8'])))
scaled_df['sensor8'] = std_scaler1.fit_transform(train_data[['sensor8']])
scaled_df['sensor9'] = std_scaler2.fit_transform(train_data[['sensor9']])
scaled_df['sensor10'] = std_scaler3.fit_transform(train_data[['sensor10']])
scaled_df['sensor419'] = r_scaler.fit_transform(train_data[['sensor419']])
scaled_df['sensor420'] = p_scaler.fit_transform(train_data[['sensor420']])
scaled_df['sensor421'] = mm_scaler.fit_transform(train_data[['sensor421']])
```

```python
print('Before Transformation')
for i in [7,8,9,418,419,420]:
    print(f'sensor{i+1}')
    _, p_val = normaltest(train_data.iloc[:,i])
    print(p_val)
print('After Transformation')
for idx, i in enumerate([7,8,9,418,419,420]):
    print(f'sensor{i+1}')
    _, p_val = normaltest(scaled_df.iloc[:,idx])
    print(p_val)
```

## Quantization (binning)

```python
from sklearn.preprocessing import KBinsDiscretizer

n_bins = 4

qbin_discretizer = KBinsDiscretizer(n_bins=n_bins, encode='ordinal', strategy='quantile')
```

```python
scaled_qbin_df = pd.DataFrame(index = range(len(train_data['sensor8'])))
scaled_qbin_df['sensor8']= qbin_discretizer.fit_transform(train_data[['sensor8']])
scaled_qbin_df['sensor9']= qbin_discretizer.fit_transform(train_data[['sensor9']])
scaled_qbin_df['sensor10']= qbin_discretizer.fit_transform(train_data[['sensor10']])
scaled_qbin_df['sensor419']= qbin_discretizer.fit_transform(train_data[['sensor419']])
scaled_qbin_df['sensor420']= qbin_discretizer.fit_transform(train_data[['sensor420']])
scaled_qbin_df['sensor421']= qbin_discretizer.fit_transform(train_data[['sensor421']])
# to analyze, we need to know this
qbin_discretizer.bin_edges_
```

```python
sensors = ['sensor8', 'sensor9', 'sensor10','sensor419', 'sensor420', 'sensor421']

for sensor in sensors:
    fig, axes = plt.subplots(ncols=2, figsize=(16,6))
    for edge in qbin_discretizer.bin_edges_[0]: # quantile bins
        line = axes[0].axvline(edge, color='r')
    axes[0].legend([line], ['Quantile bin Edges'], fontsize=10)
    sns.histplot(data=train_data, x=sensor, ax=axes[0], kde=True).set(title='Original Data') # 원본 데이터
    sns.countplot(data=scaled_qbin_df, x=sensor, ax=axes[1]).set(title='KBinsDiscretizer Data(bin = 4)') # transformation 후 데이터
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout, Flatten
from tensorflow.keras.layers import TimeDistributed
```

```python
# DataFrame을 Numpy 배열로 변환
X_train_array = X_train.values
X_test_array = X_test.values
```

```python
# 3차원 형태로 재구성
X_train_reshaped = X_train_array.reshape((X_train_array.shape[0], 1, X_train_array.shape[1]))
X_test_reshaped = X_test_array.reshape((X_test_array.shape[0], 1, X_test_array.shape[1]))
```

```python
def create_sequences(data, sequence_length):
    sequences = []
    for i in range(len(data) - sequence_length + 1):
        sequences.append(data[i:i + sequence_length])
    return np.array(sequences)

# 시퀀스 길이
sequence_length = 10

# 데이터를 시퀀스로 변환
X_train_reshaped = create_sequences(X_train_array, sequence_length)
X_test_reshaped = create_sequences(X_test_array, sequence_length)

# 결과 형태 확인
print("X_train_reshaped shape:", X_train_reshaped.shape)
print("X_test_reshaped shape:", X_test_reshaped.shape)
```

```python
#many-to-one

from keras.models import Sequential
from keras.layers import LSTM, Dropout, Dense

def create_model(units=50, dropout_rate=0.2, optimizer='adam'):

    model = Sequential()
    model.add(LSTM(units, input_shape=(10, 500)))  # return_sequences=True 제거
    model.add(Dropout(dropout_rate))
    model.add(Dense(1, activation='sigmoid'))  # TimeDistributed 레이어 제거
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model
```

```python
# 기본 모델 생성
model = create_model()
```

```python
# y_train 조정
y_train_adjusted = y_train[len(y_train) - len(X_train_reshaped):]

# 모델 학습
model.fit(X_train_reshaped, y_train_adjusted, epochs=10, batch_size=32)
```

```python
y_train_adjusted
```

```python
# 예측 수행
y_pred = model.predict(X_test_reshaped)

# 임계값을 사용하여 이진 클래스로 변환
y_pred_classes = (y_pred > 0.5).astype(int)
```

```python
# 모델 예측 값
y_pred_classes
```

```python
y_test_adjusted = y_test[-len(y_pred_classes):]

# Confusion Matrix 계산
cm = confusion_matrix(y_test_adjusted, y_pred_classes)
print("Confusion Matrix:\n", cm)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# 정확도
accuracy = accuracy_score(y_test_adjusted, y_pred_classes)
print("Accuracy:", accuracy)

# 정밀도
precision = precision_score(y_test_adjusted, y_pred_classes)
print("Precision:", precision)

# 재현율
recall = recall_score(y_test_adjusted, y_pred_classes)
print("Recall:", recall)

# F1 점수
f1 = f1_score(y_test_adjusted, y_pred_classes)
print("F1 Score:", f1)
```
